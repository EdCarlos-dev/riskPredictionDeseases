# Data Dictionary

Demographics

1. SEX (1 - Male, 2 - Female)
2. MARITAL (1 - Married, 2 - Divorced, 3 - Widowed, 4 - Separated, 5 - Never Married, 6 - A member of an unmarried couple, 9 - Refused)
3. EDUCA (1 - Never attended school, 2 - Grades 1 thr 8, 3 - Grades 9 thr 11, 4 - Grades 12 or High School graduate, 5 - College 1 year to 3 years, 6 - College 4 years or more, 9 - Refused)
4. VETERAN3 : Are you a veteran? (1 - Yes, 2 - No, 7 - Dont know, 9 - Refused, BLANK - Not Asked)
6. INCOME2 : Income Level (1 - less than 10K, 2 - Less than 15k, 3 - Less than 20k, 4 - less than 25k, 5 - less than 35k, 6 - less than 50k, 7 - less than 75k, 8 - 75k or more, 77 - dont know, 99 - Refused, BLANK - Not asked)
7. WEIGHT2 : Weight in Pounds (exception: 9000 - 9998 Weight (kilograms), 7777 Don’t know)
8. HEIGHT3 : Height in Feet and inches (exception: 9000 - 9998 Height (meters/centimeters), 7777 Don’t know )

Healthy Days: (88 is NONE, 77 is DONT KNOW, 99 is REFUSED, BLANK is NOT ASKED)

1. PHYSHLTH :  for how many days during the past 30 days was your physical health not good?   
2. MENTHLTH :  for how many days during the past 30 days was your mental health not good? 
3. POORHLTH :  for about how many days did poor physical or mental health keep you from doing your usual activities, such as self-care, work, or recreation? 

Health Care Coverage: (1 is YES, 2 is NO,  7 is DONT KNOW, 9 is REFUSED, BLANK is NOT ASKED)

1. HLTHPLN1 : Do you have any kind of health care coverage, including health insurance, prepaid plans such as HMOs, or government plans such as Medicare, or Indian Health Service?        
2. MEDCOST :  Was there a time in the past 12 months when you needed to see a doctor but could not because of cost?
3. CHECKUP1 : Routing Checkup (1 - Within past year; 2 - Within past 2 years; 3 - Within past 5 years; 4 - 5 or more years ago; 8 - Never; 9 - Refused; Blank - Not asked)

Chronic Health Conditions

1. CVDINFR4 : Ever Diagnosed with heart attack? (1 - Yes, 2 - No, 7 - Dont know, 9 - Refused)
2. CVDCRHD4 : Ever Diagnosed with Angina or Coronary Heart Disease? (1 - Yes, 2 - No, 7 - Dont know, 9 - Refused, Blank - Not asked)
3. CVDSTRK3 : Ever Diagnosed with a Stroke? (1 - Yes, 2 - No, 7 - Dont know, 9 - Refused)
4. ASTHNOW  : Still Have Asthma? (1 - Yes, 2 - No, 7 - Dont know, 9 - Refused, Blank - Not asked)
5. CHCSCNCR : Ever told that you had skin cancer? (1 - Yes, 2 - No, 7 - Dont know, 9 - Refused, Blank - Not asked)
6. CHCOCNCR : Ever told that you had any other types of cancer? (1 - Yes, 2 - No, 7 - Dont know, 9 - Refused)
7. HAVARTH3 : Told Have Arthritis? (1 - Yes, 2 - No, 7 - Dont know, 9 - Refused, Blank - Not asked)
8. ADDEPEV2 : Ever told you had a depressive disorder? (1 - Yes, 2 - No, 7 - Dont know, 9 - Refused)
9. CHCKIDNY : Ever told that you have kidney disease? (1 - Yes, 2 - No, 7 - Dont know, 9 - Refused)
10. DIABETE3 : Ever told that you have diabetes? (1 - Yes, 2 - Yes, but female told only during pregnancy, 3- No - Check SEX, 4 - No, pre-diabetes or borderline diabetes, again check SEX, 7 - Dont know, 9 - Refused, Blank - Not asked)
11. DIABAGE2 : Age When Told Diabetic? (1-97 - Age in Years, 98 - Dont Know, 99 - Refused, BLANK - Not Asked)

Tobacco Use

1. SMOKDAY2 : Frequency of Days Now Smoking (1 - Every Day, 2 - Some Day, 3 - Not at all, 7 - Dont know, 9 - Refused, BLANK - Not asked)
2. STOPSMK2 : Stopped Smoking in the past 12 months (1 - Yes, 2 - No, 7 - Don't Know, 9 - Refused, BLANK - Not asked)
3. USENOW3  : Use of Smokeless Tobacco Products (1 - Everyday, 2 - Some Days, 3 - Not at all, 7 - Dont know, 9 - Refused, BLANK - Not asked)

Alcohol Consumption

1. ALCDAY5 : Days in past 30 had alcoholic beverage (101-199 : Days per week, 201-299 : Days in past 30 days, 777 - Dont know)
2. AVEDRNK2 : Avg alcoholic drinks per day in past 30 (1-76: Number of drinks, 77 - dont know, 99 - refused, BLANK - not asked)
3. DRNK3GE5 : Binge drinking (1-76: Number of times, 88 - None, 77 - dont know, 99 - refused, BLANK - not asked)
4. _RFDRHV5 : Heavy Alcohol Consumption (1: No, 2: Yes, 9: Dont know)

Fruits & Vegetables

1. FRUIT1 : How many times did you eat fruit? (101-199: Times per day, 201-299: Times per week, 300: Less than one time per month, 301-399: Times per month, 555: Never, 777: Dont know, 999: Refused, Blank: Not asked)

Exercise

1. EXERANY2 : Exercise in Past 30 days (1: Yes, 2: No, 7: Dont know, 9: Refused, Blank: Not asked)

Calculated Variables

1. _RFHLTH : Adults with good or better health (1: Good or better, 2: Fair, 9: Dont know)
2. _HCVU651 : Respondents aged 18-64 with health care coverage (1: Have health care coverage, 2: Do not have health care coverage, 9: Dont know)
3. _RFHYPE5 : Adults who have been told they have high blood pressure by a doctor, nurse or other health professional (1: No, 2: Yes, 9: Dont know)
4. _CHOLCHK : Cholesterol Checked (1: Had cholesterol checked in past 5 years, 2: Did not have cholesterol checked, 3: Have never had cholesterol checked, 9: Dont know)
5. _RFCHOL : High Cholesterol (1: No, 2: Yes, 9: Dont know, BLANKL: Missing)
6. _MICHD : Ever had Coronary heart disease or myocardial infarction (1: Reported having MI or CHD, 2: Did not report having MI or CHD, BLANK: Not asked)
7. _CASTHM1 : Adults who currently have asthma (1: No, 2: Yes, 9: Dont know)
8. _DRDXAR1 : Respondents diagnosed with arthritis (1: Diagnosed with arthritis, 2: Not diagnosed, BLANK: Dont know)
9. _AGEG5YR : Age Category (1: 18-24, 2: 25-29, 3: 30-34, 4: 35-39, 5: 40-44, 6: 45-49, 7: 50-54, 8: 55-59, 9: 60-64, 10: 65-69, 11: 70-74, 12: 75-79, 13: 80 or older, 14: Dont know)
10. _AGE65YR: Two age groups (1: 18-64, 2: 65 or older, 3: Dont know)
11. HTIN4 : Reported height in inches (36-95: Height in inches)
12. HTM4 : Reported height in meters (91-244)
13. WTKG3 : Weight in KG (99999 - Dont know)
14. _BMI5 : BMI (1-9999: 1 or greater, BLANK: Dont know)
15. _BMI5CAT : BMI Category (1: Underweight, 2: Normal Weight, 3: Overweight, 4: Obese, Blank: Dont know)
16. _SMOKER3 : Smoking Status (1: everyday, 2: Some days, 3: Former, 4: Never, 9: Dont know)
17. _EDUCAG : Education completed category (1: Did not graduate, 2: Graduated high school, 3: Attended college or technical school, 4: graduated from college or technical school, 9: Dont know)
18. _INCOMG : Income category (1: <15k, 2: 15k to <25k, 3: 25k to <35k, 4: 35k to <50k, 5: 50k or more, 9: Dont know)
19. _PACAT1 : Physical Activity Categories (1: highly active, 2: Active, 3: Insufficiently Active, 4: Inactive, 9: Dont know)
20. _AIDTST3 : HIV (1: Yes, 2: No, 9: Dont know, BLANK: Missing)

```python
from pyspark.sql.types import StringType, IntegerType, DoubleType, BooleanType
import matplotlib.pyplot as plt
import seaborn as sns

def univariate_analysis(df):

    # Get column names and their data types
    columns = df.columns
    column_types = [field.dataType for field in df.schema.fields]

    for col_name, col_type in zip(columns, column_types):
        # Univariate analysis for numerical columns
        if isinstance(col_type, (IntegerType, DoubleType)):
            plt.figure(figsize=(12, 6))
            df.select(col_name).toPandas().plot(kind='hist', bins=20, edgecolor='black')
            plt.title(f'Univariate Analysis - {col_name}')
            plt.xlabel(col_name)
            plt.ylabel('Frequency')
            plt.show()

        # Univariate analysis for categorical columns
        elif isinstance(col_type, (StringType, BooleanType)):
            plt.figure(figsize=(12, 6))
            sns.countplot(x=col_name, data=df.toPandas())
            plt.title(f'Univariate Analysis - {col_name}')
            plt.xlabel(col_name)
            plt.ylabel('Frequency')
            plt.xticks(rotation=45, ha='right')
            plt.show() 
```

```python
def univariate_analysis_comparison(df1, df2):
    """
    Compare univariate analysis between two PySpark DataFrames.

    Parameters:
    - df1, df2 (pyspark.sql.DataFrame): Input PySpark DataFrames for the two years.

    Returns:
    - None (Plots are displayed).
    """

    # Get column names and their data types from the first DataFrame
    columns = df1.columns
    column_types = [field.dataType for field in df1.schema.fields]

    for col_name, col_type in zip(columns, column_types):
        # Univariate analysis for numerical columns
        if isinstance(col_type, (IntegerType, DoubleType)):
            plt.figure(figsize=(12, 6))
            plt.hist([df1.select(col_name).toPandas()[col_name], df2.select(col_name).toPandas()[col_name]],
                     bins=20, edgecolor='black', label=['2017', '2018'], alpha=0.7)
            plt.title(f'Univariate Analysis - {col_name} (2017 vs. 2018)')
            plt.xlabel(col_name)
            plt.ylabel('Frequency')
            plt.legend()
            plt.show()

        # Univariate analysis for categorical columns
        elif isinstance(col_type, (StringType, BooleanType)):
            plt.figure(figsize=(12, 6))
            sns.countplot(x=col_name, data=df1.toPandas(), color='blue', label='2017')
            sns.countplot(x=col_name, data=df2.toPandas(), color='orange', label='2018')
            plt.title(f'Univariate Analysis - {col_name} (2017 vs. 2018)')
            plt.xlabel(col_name)
            plt.ylabel('Frequency')
            plt.legend()
            plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
            plt.show()
```

```python
 
import matplotlib.pyplot as plt
import seaborn as sns

def plot_health_trends(df, date_column, mental_health_column, physical_health_column):
    """
    Plot trends of mental health and physical health over the past 30 days.

    Parameters:
    - df (pyspark.sql.DataFrame): Input PySpark DataFrame.
    - date_column (str): Name of the date column.
    - mental_health_column (str): Name of the mental health column.
    - physical_health_column (str): Name of the physical health column.

    Returns:
    - None (Plots are displayed).
    """

    # Assuming your DataFrame contains a date column and numerical columns for mental and physical health
    df = df.select(date_column, mental_health_column, physical_health_column).toPandas()

    # Filter data for the past 30 days
    df = df.sort_values(by=date_column, ascending=True)
    last_30_days = df.tail(30)

    # Plot trends
    plt.figure(figsize=(12, 6))
    plt.plot(last_30_days[date_column], last_30_days[mental_health_column], label='Mental Health', marker='o')
    plt.plot(last_30_days[date_column], last_30_days[physical_health_column], label='Physical Health', marker='o')
    
    plt.title('Mental and Physical Health Trends (Past 30 Days)')
    plt.xlabel('Date')
    plt.ylabel('Health Score')
    plt.xticks(rotation=45, ha='right')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()


plot_health_trends(eda_df, 'YEAR', 'BADMENTHLTH_PAST30DAYS', 'BADPHYSHLTH_PAST30DAYS')
 
```
